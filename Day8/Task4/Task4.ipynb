```
df_cell_phones=spark.read.table("cell_phones_cleaned_1_csv")
df_population=spark.read.table("pop_cleaned_1_csv")
```

```
import matplotlib.pyplot as plt
import pyspark.sql.functions as F

# Merge datasets on 'country' and 'Year'
df_merged = df_cell_phones.join(df_population, on=["country", "Year"], how="inner")

# Calculate cell phones per person
df_merged = df_merged.withColumn("Cell_Phones_Per_Person", F.col("Cell_Phones") / F.col("Population"))

# Aggregate data by year (sum total cell phones and calculate global avg per person)
df_yearly = df_merged.groupBy("Year").agg(
    F.sum("Cell_Phones").alias("Total_Cell_Phones"),
    F.avg("Cell_Phones_Per_Person").alias("Avg_Cell_Phones_Per_Person")
).orderBy("Year")

# Convert to Pandas for plotting
df_yearly_pd = df_yearly.toPandas()

# Plot
fig, ax1 = plt.subplots(figsize=(10, 6))

# Line for cell phones per person
ax1.plot(df_yearly_pd["Avg_Cell_Phones_Per_Person"], df_yearly_pd["Year"], label="Cell Phones Per Person", color="blue")
ax1.set_xlabel("Number of Cell Phones per Person")
ax1.set_ylabel("Year")
ax1.set_title("Global Rise of Cell Phones and Cell Phones Per Person")

plt.legend()
plt.grid()
plt.show()
```
