```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_excel(r"hapiscore_whr.xlsx")

# Set the country column as the index and transpose the dataframe
df.set_index("country", inplace=True)
df = df.transpose()

# Plot each country's happiness trend over the years
plt.figure(figsize=(7, 5))  

for country in df.columns:
    plt.plot(df.index, df[country], marker="o", linestyle="--", linewidth=1, label=country)

# Labels and title
plt.ylabel("Year", fontsize=12)  
plt.xlabel("Years", fontsize=12)  # Fixed the error here
plt.title("Happiness Over Time", fontsize=14, fontweight="bold")

# Grid and legend
plt.grid(True, linestyle="--", alpha=0.5)
plt.gca().invert_yaxis()  # Keep the most recent year at the top
plt.legend(loc="best", fontsize=8)

plt.tight_layout()
plt.show()
```
