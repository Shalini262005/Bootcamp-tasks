```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("C:/Users/shali/Downloads/lex.xlsx")
df.set_index("country", inplace=True)
df = df.transpose()  
df
```

```python
df["Average"] = df.mean(axis=1)
df
```

```python
plt.figure(figsize=(5,5))
plt.plot(df["Average"], df.index, marker="o", linestyle="--", color="GREEN", linewidth=1, label="Average")
plt.ylabel("Year", fontsize=12)  
plt.xlabel("Average Life Expectancy (Years)", fontsize=12) 
plt.title(" Life Expectancy Trends Over Time", fontsize=14, fontweight="bold")
plt.grid(True, linestyle="--", alpha=0.5)
plt.gca() 
plt.legend()
plt.tight_layout()
plt.show()
```
