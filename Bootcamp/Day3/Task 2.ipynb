```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_excel("C:/Users/shali/Downloads/data.xlsx")
df.set_index("country",inplace=True)
df=df.transpose()
df
df["Average"]=df.mean(axis=1)
plt.figure(figsize=(5,5))
plt.plot(df["Average"],df.index,marker="o",linestyle="--",color="Blue",linewidth=1,label="Average")
plt.ylabel("Year",fontsize=12)
plt.xlabel("Average Income",fontsize=12)
plt.title("Average Income over time",fontsize=14,fontweight="bold")
plt.grid(True,linestyle="--",alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```
