```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df2=pd.read_csv("C:/Users/shali/OneDrive/Bootcamp/Day7/Task1/cell_phones_total.csv")
df2
```

```python
df2=df2.drop(columns=["1960","1961","1962","1963","1964","1965","1966","1967","1968","1969","1970","1971","1972","1973","1974","1975","1976","1977","1978","1979"],axis=1)
df2
```

```python
def convert_to_numeric(value):
    if isinstance(value, str):  # Ensure it's a string before processing
        if 'B' in value:
            return float(value.replace('B', '')) * 1_000_000_000
        elif 'M' in value:
            return float(value.replace('M', '')) * 1_000_000
        elif 'k' in value:
            return float(value.replace('k', '')) * 1_000
        else:
            return float(value)  # Convert numeric strings directly to float
    return value  # Return unchanged if already numeric


# Convert all non-country columns to numeric values
df2.iloc[:, 1:] = df2.iloc[:, 1:].astype(str).applymap(convert_to_numeric)

```

```python
df2.duplicated().sum()
```

```python
df2.fillna(0, inplace=True)
df2
```

```python
# Convert wide format to long format
df2 = df2.melt(id_vars=['country'], var_name='Year', value_name='Cell Phones')
df2
```

```python
df2.to_csv("cell_phones_cleaned.csv", index=False)

print("Data cleaning complete! File saved as 'cell_phones_cleaned.csv")
```

```python
df1=pd.read_csv("C:/Users/shali/OneDrive/Bootcamp/Day7/Task1/pop.csv")
```

```python
df1 = df1.melt(id_vars=['country'], var_name='Year', value_name='Population')

# Convert 'Year' to integer
df1['Year'] = df1['Year'].astype(int)
df1
```

```python
def convert_units(value):
    if isinstance(value, str):
        if 'B' in value:
            return float(value.replace('B', '')) * 1_000_000_000
        elif 'M' in value:
            return float(value.replace('M', '')) * 1_000_000
        elif 'k' in value:
            return float(value.replace('k', '')) * 1_000
        else:
            return float(value)  # Convert numeric strings directly to float
    return value
df1['Population'] = df1['Population'].apply(convert_units)
df1
```

```python
df1['Population'] = df1['Population'].interpolate() 
```
